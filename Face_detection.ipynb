# opencv

```
import cv2 # importing opencv
```

```
 model=cv2.CascadeClassifier("/content/haarcascade_frontalface_default.xml") #reading the classifier
```

```
faces=model.detectMultiScale(pic)
```

```
faces
```

```
faces.shape
```

```
# the faces are (5,4) whih means the picture contain 5 faces
# and each face has for different properties
```

```
for face in faces:
  x,y,w,h = face
  pic = cv2.rectangle(pic,(x,y),(x+w,y+h),(0,200,100),3)
```

```
plt.imshow(pic)
```
